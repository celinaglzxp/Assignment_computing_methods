# Part III: Functions
### 1) Write a function to convert temperature from kelvin to celsius

```python
def kelvin_to_celsius(K):
    return K - 273.15

def celsius_to_kelvin(C):
    return C + 273.15
```

```python
kelvin_to_celsius(300)
```

```python
celsius_to_kelvin(27)
```

### 2) Write a function to convert temperature to fahrenheit

```python
def fahrenheit(T, kelvin=False):
    if kelvin:
        C = kelvin_to_celsius(T)
    else:
        C = T
    return C * 1.8 + 32
```

### 3) Check that the outputs are sensible

```python
fahrenheit(27)
```

```python
fahrenheit(300, kelvin=True)
```

```python
fahrenheit(0)
```

### 4) Now write a function that converts from farenheit

```python
def F_to_T(F, kelvin=False):
    C = (F - 32) / 1.8
    if kelvin:
        return celsius_to_kelvin(C)    # C + 273.15
    else:
        return C
```

```python
F_to_T(80)
```

```python
F_to_T(80, kelvin=True)
```

### 5) Write a function that takes two arguments (feet and inches) and returns height in meters

```python
def height_M(feet, inches):
    return (feet * 12 + inches) * 0.0254
```

```python
height_M(5,5)
```

```python
height_M(5,7)
```

```python
height_M(6,3)
```

### 6. Write a function takes one argument (height in meters) and returns two arguments (feet and inches)

```python
def height_in(M):
    h = M / 0.0254
    feet = int(h // 12)
    # use int to align with the normal way to write height in feet and inches
    inches = int(h - feet * 12)
    return feet, inches
```

```python
height_in(1.65)
```

```python
height_in(2.0)
```

### 7) Verify that the “round trip” conversion from and back to meters is consistent

```python
test = [1.65, 1.85, 2.00]

for M in test:
    feet, inches = height_in(M)
    M_back = height_M(feet, inches)
    print(M, feet, inches, M_back)   
    # int() rounds inches down so M_back may differ slightly from M
```
